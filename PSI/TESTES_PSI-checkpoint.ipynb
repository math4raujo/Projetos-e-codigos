```python
import pandas as pd
import numpy as np
import math
```

```python
#importando as bases
df_1 = pd.read_csv('train.csv')
df_2 = pd.read_csv('test.csv')

df_treino
#df_teste.columns

selecionadas = ['age', 'gender', 'avg_transaction_value', 'avg_time_spent', 'used_special_discount', 'offer_application_preference', 'past_complaint']

df_treino = df_1[selecionadas]
df_teste = df_2[selecionadas]

df_treino
```

```python
def transforma_vars(df):
    
    n = len(df)
    
    for i in range(n):
    
        if df.loc[i, 'gender'] == 'F':
            
            df.loc[i, 'gender'] = 1
    
        elif df.loc[i, 'gender'] == 'M':
            
            df.loc[i, 'gender'] = 0
    
        else:
            
            df.loc[i, 'gender'] = ''
            
        if df.loc[i, 'used_special_discount'] == 'Yes':
            
            df.loc[i, 'used_special_discount'] = 1
        else:
            
            df.loc[i, 'used_special_discount'] = 0
            
        
        if df.loc[i, 'offer_application_preference'] == 'Yes':
            
            df.loc[i, 'offer_application_preference'] = 1
        
        else:
            
            df.loc[i, 'offer_application_preference'] = 0
            
            
        if df.loc[i, 'past_complaint'] == 'Yes':
            
            df.loc[i, 'past_complaint'] = 1
        
        else:
            
            df.loc[i, 'past_complaint'] = 0 
            
```

```python
transforma_vars(df_teste)
```

```python
def cria_classes_dummy(df, var):
    
    n = len(df)
    
    conta_um = 0
    conta_zero = 0
    
    for i in range(n):
        if df.loc[i, var] == 1:
            conta_um += 1
        else:
            conta_zero += 1
    
    newDF = pd.DataFrame()
    newDF.loc[0,'VAR_NAME'] = var
    newDF.loc[0, 'contaZero'] = int(conta_zero)
    newDF.loc[0, 'contaUm'] = int(conta_um)
    
    newDF.loc[1, 'contaZero'] = conta_zero/(conta_um + conta_zero)
    newDF.loc[1, 'contaUm'] = conta_um/(conta_um + conta_zero)
    
    return newDF
```

```python
#esperado
df3 = cria_classes_dummy(df_treino, 'past_complaint')
df3
```

```python
#observado
df4 = cria_classes_dummy(df_teste, 'past_complaint')
df4
```

```python
# CALCULO DO PSI

faixa1 = (df4.loc[1, 'contaZero'] - df3.loc[1, 'contaZero']) * np.log(df4.loc[1, 'contaZero']/df3.loc[1, 'contaZero'])
faixa2 = (df4.loc[1, 'contaUm'] - df3.loc[1, 'contaUm']) * np.log(df4.loc[1, 'contaUm']/df3.loc[1, 'contaUm'])

faixa1 + faixa2
```
